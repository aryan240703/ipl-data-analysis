```python
import pandas as pd
df= pd.read_csv('IPL_Matches_2008_2022.csv')
df
```

```python
#import all lib.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#convert data columns to datatime
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
```

```python
df = df.dropna(subset=['WinningTeam'])   #drop rows which missing
df
```

```python
# Check data types and missing values
print(df.info())
print(df.isnull().sum())
```

```python
# Most successful teams
sns.countplot(data=df, y='WinningTeam', order=df['WinningTeam'].value_counts().index)
plt.title('Most Wins by Teams')
plt.show()
```

```python
# Toss Decision Impact
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='TossDecision', hue='WinningTeam')
plt.title("Match Wins Based on Toss Decision")
plt.xlabel("Toss Decision")
plt.ylabel("Number of Matches Won")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Venue Analysis — Most Frequent Match Winners per Venue
top_venues = df['Venue'].value_counts().head(10).index
venue_df = df[df['Venue'].isin(top_venues)]

plt.figure(figsize=(14,6))
sns.countplot(data=venue_df, y='Venue', hue='WinningTeam', palette='Set2')
plt.title("Match Wins by Teams at Top 10 Venues")
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.tight_layout()
plt.show()
```

```python
 #Player of the Match — Top 10
top_players = df['Player_of_Match'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_players.values, y=top_players.index, palette='magma')
plt.title("Top 10 Players with Most Player of the Match Awards")
plt.xlabel("Number of Awards")
plt.ylabel("Player")
plt.tight_layout()
plt.show()
```

```python
#Batting First vs Chasing — Match Outcomes
# Create win_type column
df['WinBy'] = df['WonBy'].fillna('Unknown')
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='WinBy')
plt.title("Matches Won by Batting vs Chasing")
plt.xlabel("Win Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
